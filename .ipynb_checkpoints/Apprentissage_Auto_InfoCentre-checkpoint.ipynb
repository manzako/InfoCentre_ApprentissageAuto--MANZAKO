### (7) Séparez l’ensemble de données résultant (de la segmentation) en données d’apprentissage et en données de test ; utilisez les méthodes suivantes pour construire vos modèles prédictifs

```python
seed=70
Y=DonneeVin['Etiquete']
X=DonneeVin.drop(['Etiquete'],axis=1)

X_train,X_test,Y_train,Y_test=train_test_split(X,Y, test_size=0.40,random_state=seed, stratify=Y)
print(X_train.shape)
print(X_test.shape)
```

##### Regression logistique

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
lr = LogisticRegression(random_state=seed)
lr_hyper={'C':[0.001, 0.01, 0.1, 1, 10, 100, 1000]}
lr_cv=GridSearchCV(lr, lr_hyper, cv=5)
lr_cv.fit(X_train,Y_train)
print(lr_cv.best_score_)
print(lr_cv.best_estimator_)
```

##### Arbres de decision (methode CART ET i+ID3)

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import ShuffleSplit
#Déclaration de l'arbre de décision
clf_tree = DecisionTreeClassifier()
cv = ShuffleSplit(n_splits=30, test_size=0.2)
depth_grid = [1, 3, 5, 10, 20, 30, 40, 50, None]
parameters = {'max_depth': depth_grid} 
clf_tree_grid = GridSearchCV(clf_tree, parameters, cv=cv)
clf_tree_grid.fit(X_train, Y_train)
print(clf_tree_grid.best_score_)
print('Meilleure valeur du paramètre: ', clf_tree_grid.best_params_)
```

##### K-plus proches voisin

```python
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.neighbors import KNeighborsClassifier


KNN=KNeighborsClassifier()
KNN_hyper={
    'n_neighbors':[1,10,20,30]
}
KNN_CV=GridSearchCV(KNN,KNN_hyper,cv=5)
KNN_CV.fit(X_train,Y_train)
print(KNN_CV.best_score_)
print(KNN_CV.best_estimator_)
```

### (8) À l’aide des données de test, discutez de l’efficacité de chacune des méthodes utilisées 

```python
def Evaluation_modele(model,features,labels):
    pred=model.predict(features)
    score=accuracy_score(Y_test,pred) # Nota: chaque modele a sa metrique
    print('Score  :', round(score,3))
    
models=[lr_cv.best_estimator_, 
        KNN_CV.best_estimator_,
        clf_tree_grid.best_estimator_]
for model in models:
    print('Model :'+str(model))
    Evaluation_modele(model,X_test, Y_test)
    
    print('-'*90)
```
